```python
import json
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def set_axis_style(ax, labels):
    ax.set_xticks(np.arange(1, len(labels) + 1), labels=labels)
    ax.set_xlim(0.25, len(labels) + 0.75)
    # ax.set_xlabel('Sample name')
```

```python
folder_location = '../results/hashes/'

original = json.load(open(f'{folder_location}original.json'))
modern = json.load(open(f'{folder_location}modern.json'))

opt_first_line = json.load(open(f'{folder_location}opt_first_line.json'))
opt_256_tokens = json.load(open(f'{folder_location}opt_256_tokens.json'))
opt_512_tokens = json.load(open(f'{folder_location}opt_512_tokens.json'))

llama_first_line = json.load(open(f'{folder_location}llama_first_line.json'))
llama_256_tokens = json.load(open(f'{folder_location}llama_256_tokens.json'))
llama_512_tokens = json.load(open(f'{folder_location}llama_512_tokens.json'))

alpaca_1 = json.load(open(f'{folder_location}alpaca_1.json'))
alpaca_3 = json.load(open(f'{folder_location}alpaca_3.json'))

alpaca_free_2 = json.load(open(f'{folder_location}alpaca_free_2.json'))
alpaca_free_4 = json.load(open(f'{folder_location}alpaca_free_4.json'))

original = Counter(original)
modern = Counter(modern)

opt_first_line = Counter(opt_first_line)
opt_256_tokens = Counter(opt_256_tokens)
opt_512_tokens = Counter(opt_512_tokens)

llama_first_line = Counter(llama_first_line)
llama_256_tokens = Counter(llama_256_tokens)
llama_512_tokens = Counter(llama_512_tokens)

alpaca_1 = Counter(alpaca_1)
alpaca_3 = Counter(alpaca_3)

alpaca_free_2 = Counter(alpaca_free_2)
alpaca_free_4 = Counter(alpaca_free_4)

def normalize_counter(d):
    total = sum(d.values())
    new_counter = Counter()
    for k in d:
        new_counter[k] = (d[k]/total) * 100
    return new_counter

original = normalize_counter(original)
modern = normalize_counter(modern)

opt_first_line = normalize_counter(opt_first_line)
opt_256_tokens = normalize_counter(opt_256_tokens)
opt_512_tokens = normalize_counter(opt_512_tokens)

llama_first_line = normalize_counter(llama_first_line)
llama_256_tokens = normalize_counter(llama_256_tokens)
llama_512_tokens = normalize_counter(llama_512_tokens)

alpaca_1 = normalize_counter(alpaca_1)
alpaca_3 = normalize_counter(alpaca_3)

alpaca_free_2 = normalize_counter(alpaca_free_2)
alpaca_free_4 = normalize_counter(alpaca_free_4)
```

```python
print('Original and Modern')
print(sum((original & modern).values()))

print('Original and generation')
print(sum((original & opt_first_line).values()))
print(sum((original & opt_256_tokens).values()))
print(sum((original & opt_512_tokens).values()))

print(sum((original & llama_first_line).values()))
print(sum((original & llama_256_tokens).values()))
print(sum((original & llama_512_tokens).values()))

print(sum((original & alpaca_1).values()))
print(sum((original & alpaca_free_2).values()))
print(sum((original & alpaca_3).values()))
print(sum((original & alpaca_free_4).values()))

print('Modern and generation')
print(sum((modern & opt_first_line).values()))
print(sum((modern & opt_256_tokens).values()))
print(sum((modern & opt_512_tokens).values()))

print(sum((modern & llama_first_line).values()))
print(sum((modern & llama_256_tokens).values()))
print(sum((modern & llama_512_tokens).values()))

print(sum((modern & alpaca_1).values()))
print(sum((modern & alpaca_free_2).values()))
print(sum((modern & alpaca_3).values()))
print(sum((modern & alpaca_free_4).values()))
```
