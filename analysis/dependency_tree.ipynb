```python
import json
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def set_axis_style(ax, labels):
    ax.set_xticks(np.arange(1, len(labels) + 1), labels=labels)
    ax.set_xlim(0.25, len(labels) + 0.75)
    # ax.set_xlabel('Sample name')
```

```python
folder_location = '../results/hashes/'

original = json.load(open(f'{folder_location}original.json'))

opt_first_line = json.load(open(f'{folder_location}opt_first_line.json'))
opt_256_tokens = json.load(open(f'{folder_location}opt_256_tokens.json'))
opt_512_tokens = json.load(open(f'{folder_location}opt_512_tokens.json'))

llama_first_line = json.load(open(f'{folder_location}llama_first_line.json'))
llama_256_tokens = json.load(open(f'{folder_location}llama_256_tokens.json'))
llama_512_tokens = json.load(open(f'{folder_location}llama_512_tokens.json'))

alpaca_1 = json.load(open(f'{folder_location}alpaca_1.json'))
alpaca_3 = json.load(open(f'{folder_location}alpaca_3.json'))

alpaca_free_2 = json.load(open(f'{folder_location}alpaca_free_2.json'))
alpaca_free_4 = json.load(open(f'{folder_location}alpaca_free_4.json'))
```

```python
original = Counter(original)

opt_first_line = Counter(opt_first_line)
opt_256_tokens = Counter(opt_256_tokens)
opt_512_tokens = Counter(opt_512_tokens)

llama_first_line = Counter(llama_first_line)
llama_256_tokens = Counter(llama_256_tokens)
llama_512_tokens = Counter(llama_512_tokens)

alpaca_1 = Counter(alpaca_1)
alpaca_3 = Counter(alpaca_3)

alpaca_free_2 = Counter(alpaca_free_2)
alpaca_free_4 = Counter(alpaca_free_4)
```

```python
print(sum(original.values()), len(original))
print(sum(opt_first_line.values()), len(opt_first_line))
print(sum(opt_256_tokens.values()), len(opt_256_tokens))
print(sum(opt_512_tokens.values()), len(opt_512_tokens))
print(sum(llama_first_line.values()), len(llama_first_line))
print(sum(llama_256_tokens.values()), len(llama_256_tokens))
print(sum(llama_512_tokens.values()), len(llama_512_tokens))
print(sum(alpaca_1.values()), len(alpaca_1))
print(sum(alpaca_free_2.values()), len(alpaca_free_2))
print(sum(alpaca_3.values()), len(alpaca_3))
print(sum(alpaca_free_4.values()), len(alpaca_free_4))
```
